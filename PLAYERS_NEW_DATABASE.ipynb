```python
import pandas as pd

def obtener_datos_unicos(df, columnas):
    """
    Filtra un DataFrame para que solo contenga filas únicas basadas en una combinación de columnas.
    
    :param df: DataFrame de pandas.
    :param columnas: Lista de nombres de columnas a considerar para unicidad.
    :return: DataFrame con filas únicas según las columnas especificadas.
    """
    return df.drop_duplicates(subset=columnas)

def combinar_bases(df1, columnas_df1, df2, columnas_df2):
    """
    Une df1 con df2 basado en una clave específica y agrega las columnas únicas de df2 sin duplicar las existentes en df1.
    
    :param df1: Primer DataFrame base.
    :param columnas_df1: Lista de columnas clave en df1.
    :param df2: Segundo DataFrame con información adicional.
    :param columnas_df2: Lista de columnas clave en df2.
    :return: DataFrame combinado con columnas de df2 añadidas sin duplicar las ya existentes en df1.
    """
    # Asegurar que las claves son listas
    if isinstance(columnas_df1, str):
        columnas_df1 = [columnas_df1]
    if isinstance(columnas_df2, str):
        columnas_df2 = [columnas_df2]
    
    # Verificar que las columnas existen en los DataFrames
    for col in columnas_df1:
        if col not in df1.columns:
            raise KeyError(f"La columna {col} no está en df1")
    for col in columnas_df2:
        if col not in df2.columns:
            raise KeyError(f"La columna {col} no está en df2")
    
    # Realizar la unión
    df_comb = df1.merge(df2, left_on=columnas_df1, right_on=columnas_df2, how='left')
    
    # Obtener columnas finales sin duplicados
    columnas_finales = list(df1.columns) + [col for col in df2.columns if col not in columnas_df2 and col not in df1.columns]
    
    # Filtrar solo las columnas existentes
    columnas_finales = [col for col in columnas_finales if col in df_comb.columns]
    
    return df_comb[columnas_finales]

def seleccionar_columnas(df, columnas):
    """
    Devuelve un DataFrame con solo las columnas especificadas.
    
    :param df: DataFrame de pandas.
    :param columnas: Lista de nombres de columnas a seleccionar.
    :return: DataFrame con solo las columnas especificadas.
    """
    columnas_validas = [col for col in columnas if col in df.columns]
    return df[columnas_validas]
```

```python
def filtrar_por_valores(df, columna, valores):
    """
    Filtra un DataFrame para que solo contenga filas donde la columna especificada tenga valores en la lista dada.
    
    :param df: DataFrame de pandas.
    :param columna: Nombre de la columna a filtrar.
    :param valores: Lista de valores a mantener en la columna especificada.
    :return: DataFrame filtrado.
    """
    if columna not in df.columns:
        raise KeyError(f"La columna '{columna}' no está en el DataFrame")
    return df[df[columna].isin(valores)]

year=1938
df_players=filtrar_por_valores(df_players_global,'Year',[year])
df_players
```
